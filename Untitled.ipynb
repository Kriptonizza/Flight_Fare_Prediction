```python
f=[]
for i in range(len(dur)):
    if len(dur[i].split())==2:
        
        h=int(dur[i].split()[0].strip("h"))
        m=int(dur[i].split()[1].strip("m"))
        f.append(h*60+m)
    elif len(dur[i].split())!=2:
        if "m"==dur[i].split()[0][-1]:
            m=int(dur[i].split()[0].strip("m"))
            h=0
            f.append(h*60+m)
        else:
            m=0
            h=int(dur[i].split()[0].strip("h"))
            f.append(h*60+m)

            
     
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x=df.drop("Price",axis=1)
y=df["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
li=LinearRegression()
li.fit(x_train,y_train)
y_pred=li.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
hyp={'alpha':np.arange(0.01,3,0.01)}
rg=Ridge()
GR=GridSearchCV(rg,hyp,cv=5)
GR.fit(x_train,y_train)
GR.best_estimator_
```

```python
Rg=GR.best_estimator_
```

```python
Rg.fit(x_train,y_train)
y_pred=Rg.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
hyp={'alpha':np.arange(0.01,3,0.01)}
rg=Lasso()
GR=GridSearchCV(rg,hyp,cv=5)
GR.fit(x_train,y_train)
GR.best_estimator_
```

```python
Rg=GR.best_estimator_
```

```python
Rg.fit(x_train,y_train)
y_pred=Rg.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
nm=ExtraTreesRegressor()
nm.fit(x_train,y_train)
```

```python
y_pred=nm.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
y_pred=nm.predict(x_train)
print(mean_absolute_error(y_train,y_pred))
print(r2_score(y_train,y_pred))
```
